```python
# Realizando o Import das Blibiotecas.
import pandas as pd
import numpy as np
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
```

```python
# Carregando os dados no Objeto Pandas DATAFrame.
data = pd.read_csv("train.csv")
```

```python
# Exibindo as 5 primeiras linhas do dataframe.
data.head()
```

```python
# Exibindo as ultimas linhas do dataframe.
data.tail()
```

```python
# Separando os dados em Features de Seleção e de Classificação.
X = data.iloc[:,0:20]  #independent columns
y = data.iloc[:,-1]    #target column i.e price range
```

```python
#apply SelectKBest class to extract top 10 best features
bestfeatures = SelectKBest(score_func=chi2, k=10) # Modelo estatistico e a quantidade de colunas 
fit = bestfeatures.fit(X,y) # Treinamento do modelo
dfscores = pd.DataFrame(fit.scores_) # Mostrar os scores
dfcolumns = pd.DataFrame(X.columns) # Mostrar as colunas
```

```python
# Exibindo o resultado da Classificação.
#concat two dataframes for better visualization 

featureScores = pd.concat([dfcolumns,dfscores],axis=1)
featureScores.columns = ['Specs','Score']  #naming the dataframe columns

print(featureScores.nlargest(10,'Score'))  #print 10 best features
# As 4 primeiras são as mais importantes
```
